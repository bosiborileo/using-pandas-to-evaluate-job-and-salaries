```python
#import modules
import pandas as pd
import matplotlib.pyplot as plt
#read dataset
data = pd.read_csv('/content/drive/MyDrive/archive/jobs_in_data.csv')
```

```python
total_entries = len(data)
print("Total number of entries in the database:", total_entries)
```

```python
# Checking for missing values
missing_values = data.isnull().sum().sum()
if missing_values > 0:
    print("There are missing values in the dataset.")
    # Dropping records containing missing values
    data.dropna(inplace=True)
    print("Records with missing values have been dropped.")
else:
    print("There are no missing values in the dataset.")
```

```python
# Get information about the dataset
info = data.info()

# Get statistical summary of the dataset
description = data.describe()
print("### Dataset Information:\n\n", info)
print("\n\n### Dataset Description:\n\n", description)
```

All columns have appropriate data types, with `int64` for numerical columns and `object` for categorical columns. There are no missing values in the dataset, indicating a complete and clean dataset.

### Dataset Description:

The statistical summary of numerical columns reveals the following insights:

1. **work_year**: The majority of data entries are from the year 2023, with a few records from 2020.
2. **salary**: Salaries range from 14,000 to 450,000, with a mean salary of approximately 149,928.
3. **salary_in_usd**: Similarly, salaries in USD range from 15,000 to 450,000, with a mean salary of approximately 150,299.

The salary data appears to be normally distributed, with the mean and median being relatively close together, indicating a symmetrical distribution. This suggests that the dataset covers a wide range of job positions with varying salary levels.

Overall, the dataset provides comprehensive information about job positions, including salary details, employment attributes, and other relevant information, without any missing values, making it suitable for analysis and modeling tasks.

```python

#pie chart showing the most common job titles
job_title_counts = data['job_title'].value_counts().head(10)
plt.figure(figsize=(8, 8))
plt.pie(job_title_counts, labels=job_title_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Most Common Job Titles')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Filter the DataFrame for "Data Scientist" job title
data_scientist_df = data[data['job_title'] == 'Data Scientist']

# Save and export the filtered DataFrame as CSV
data_scientist_df.to_csv('data_scientist_jobs.csv', index=False)
```

```python
# Group the data by experience level and calculate total, min, max, and average salaries
salary_stats_by_experience = data.groupby('experience_level')['salary'].agg(['sum', 'min', 'max', 'mean']).reset_index()

# Print the statistics
print(salary_stats_by_experience)

# Visualize the differences
plt.figure(figsize=(10, 6))

# Plot total salary by experience level
plt.subplot(2, 2, 1)
plt.bar(salary_stats_by_experience['experience_level'], salary_stats_by_experience['sum'])
plt.title('Total Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Total Salary')

# min salary by experience level
plt.subplot(2, 2, 2)
plt.bar(salary_stats_by_experience['experience_level'], salary_stats_by_experience['min'])
plt.title('Minimum Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Minimum Salary')

# max salary by experience level
plt.subplot(2, 2, 3)
plt.bar(salary_stats_by_experience['experience_level'], salary_stats_by_experience['max'])
plt.title('Maximum Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Maximum Salary')

# average salary by experience level
plt.subplot(2, 2, 4)
plt.bar(salary_stats_by_experience['experience_level'], salary_stats_by_experience['mean'])
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary')

plt.tight_layout()
plt.show()
```

```python

#Filter the dataset for employees in the United Kingdom
uk_df = data[data['employee_residence'] == 'United Kingdom']

#Group the filtered dataset by employment type
employment_type_groups = uk_df.groupby('employment_type')

#Calculate summary statistics for salaries within each employment type group
salary_stats_by_employment_type = employment_type_groups['salary'].describe()

# Visualize the salary comparisons
plt.figure(figsize=(10, 6))

# Box plot for salary comparisons
salary_boxplot = uk_df.boxplot(column='salary', by='employment_type', patch_artist=True)
plt.title('Salary Comparisons by Employment Type (United Kingdom)')
plt.suptitle('')  # Suppress the default title

# Customizing plot
plt.xlabel('Employment Type')
plt.ylabel('Salary')
plt.xticks([1, 2, 3], ['Full-time', 'Part-time', 'Contract'])

plt.show()

# Print summary statistics
print("Summary Statistics for Salaries by Employment Type in the United Kingdom:")
print(salary_stats_by_employment_type)
```
